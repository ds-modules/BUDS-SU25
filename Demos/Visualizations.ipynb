# 4. Bar Charts

```python
top_movies
```

```python
sorted_by_gross = top_movies.sort('Gross', descending = True)
sorted_by_gross
```

```python
top_per_studio = sorted_by_gross.sort('Studio', distinct = True)
top_per_studio
```

```python
top_per_studio.barh('Studio', 'Gross')
```

### How would I get this sorted in *ascending* order?

```python
top_studios = top_per_studio.sort('Gross')
top_studios
```

```python
top_studios.barh('Studio', 'Gross')
```
